# 7.13 Reshaping and Transposing 

### `reshape` vs. `resize` 

The difference between reshape() and resize() is that reshape() does not change the original array but only returns the changed array, whereas the resize() method returns nothing and directly changes the original array. 

```python
import numpy as np
```

```python
grades = np.array([[87, 96, 70], [100, 87, 90]])
```

```python
grades
```

```python
grades.reshape(1, 6)
```

```python
grades
```

```python
grades.resize(1, 6)
```

```python
grades
```

### `flatten` vs. `ravel` 

```python
grades = np.array([[87, 96, 70], [100, 87, 90]])
```

```python
grades
```

```python
flattened = grades.flatten()
```

```python
flattened
```

```python
grades
```

```python
flattened[0] = 100
```

```python
flattened
```

```python
grades
```

```python
raveled = grades.ravel()
```

```python
raveled
```

```python
grades
```

```python
raveled[0] = 100
```

```python
raveled
```

```python
grades
```
