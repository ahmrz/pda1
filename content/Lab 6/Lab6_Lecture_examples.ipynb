# 7.6 List vs. `array` Performance: Introducing `%timeit` 

### Timing the Creation of a List Containing Results of 6,000,000 Die Rolls 

````python
import random
````

````python
%timeit rolls_list = \
   [random.randrange(1, 7) for i in range(0, 6_000_000)]
````

### Timing the Creation of an `array` Containing Results of 6,000,000 Die Rolls  

````python
import numpy as np
````

````python
%timeit rolls_array = np.random.randint(1, 7, 6_000_000)
````

### 60,000,000 and 600,000,000 Die Rolls  

````python
%timeit rolls_array = np.random.randint(1, 7, 60_000_000)
````

````python
%timeit rolls_array = np.random.randint(1, 7, 600_000_000)
````

### Customizing the %timeit Iterations  

Documentation: https://ipython.org/ipython-doc/dev/interactive/magics.html#magic-timeit


````python
%timeit -n3 -r2 rolls_array = np.random.randint(1, 7, 6_000_000)
````


# 7.6 Self Check

**1. _(IPython Session)_** Use `%timeit` to compare the execution time of the following two statements. The first uses a list comprehension to create a list of the integers from 0 to 9,999,999 then totals them with the built-in `sum` function. The second statement does the same thing using an `array` and its `sum` method.
```python
sum([x for x in range(10_000_000)])
np.arange(10_000_000).sum()
```

**Answer:** 

````python
import numpy as np
````

````python
%timeit sum([x for x in range(10_000_000)])
````

````python
%timeit np.arange(10_000_000).sum()
````

The statement with the list comprehension took 26 times longer to execute than the one with the `array`.
