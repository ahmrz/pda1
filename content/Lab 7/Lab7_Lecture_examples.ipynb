# 8.2.2 Field Widths and Alignment

# 8.2.3 Numeric Formatting

### Formatting Positive Numbers with Signs

````python
f'[{27:+10d}]'
````

* `+` → always show the sign (`+` for positive, `-` for negative)
* `0` → pad with zeros instead of spaces
* `10` → total width is `10` characters
* `d` → decimal integer
* Numbers are right-aligned by default, and with `0` padding the zeros come after the sign

````python
f'[{27:+010d}]'
````

### Using a Space Where a + Sign Would Appear in a Positive Value

````python
print(f'{27:d}\n{27: d}\n{-27: d}')
````

### Grouping Digits

````python
f'{12345678:,d}'
````

````python
f'{123456.78:,.2f}'
````

### Other Predefined Character Classes

### Custom Character Classes

````python
#[A-Z][a-z] --> match any uppercase ASCII letter, match any lowercase ASCII letter
#re* --> matches 0 or more occurrences of preceding expression 
#Will output 'Valid' because it contains both uppercase and lowercase letters 
'Valid' if re.fullmatch('[A-Z][a-z]*', 'Wally') else 'Invalid'
````

````python
#[A-Z][a-z] --> match any uppercase ASCII letter, match any lowercase ASCII letter
#re* --> matches 0 or more occurrences of preceding expression 
#Will output 'Invalid' because it contains lowercase letters only
'Valid' if re.fullmatch('[A-Z][a-z]*', 'eva') else 'Invalid'
````

````python
#[^...] --> Matches any single character not in brackets
'Match' if re.fullmatch('[^a-z]', 'A') else 'No match'
````

````python
#[^...] --> Matches any single character not in brackets
'Match' if re.fullmatch('[^a-z]', 'a') else 'No match'
````

````python
#[*+$] --> Match any single element *, +, and $ 
'Match' if re.fullmatch('[*+$]', '*') else 'No match'
````

````python
#[*+$] --> Match any single element *, +, and $ 
'Match' if re.fullmatch('[*+$]', '!') else 'No match'
````

### * vs. + Quantifier

````python
#[A-Z][a-z] --> match any uppercase ASCII letter, match any lowercase ASCII letter
#re+ --> matches 1 or more occurrences of preceding expression 
#Will output 'Valid' because it contains both uppercase and lowercase letters
'Valid' if re.fullmatch('[A-Z][a-z]+', 'Wally') else 'Invalid'
````

````python
#[A-Z][a-z] --> match any uppercase ASCII letter, match any lowercase ASCII letter
#re+ --> matches 1 or more occurrences of preceding expression 
#Will output 'Valid' because it contains both uppercase and lowercase letters
'Valid' if re.fullmatch('[A-Z][a-z]+', 'E') else 'Invalid'
````

#### Example showing the difference between * and + 

````python
#re* --> matches 0 or more occurrences of preceding expression 
'Valid' if re.fullmatch('[A-Z][a-z]*', 'E') else 'Invalid'
````

````python
#re+ --> matches 1 or more occurrences of preceding expression 
'Valid' if re.fullmatch('[A-Z][a-z]+', 'E') else 'Invalid'
````

# 8.12.2 Replacing Substrings and Splitting Strings

### Function sub—Replacing Patterns 

````python
import re
````

````python
#replace all occurrences of pattern \t in the string by , 
re.sub(r'\t', ', ', '1\t2\t3\t4')
````

````python
#replace all occurrences of pattern \t in the string by , only two times 
re.sub(r'\t', ', ', '1\t2\t3\t4', count=2)
````

### Function split 

````python
# ,\s* --> a pattern that matches the comma , and matches whitespace 0 or more times, and splits by this pattern 
re.split(r',\s*', '1,  2,  3,4,    5,6,7,8')
````

````python
# ,\s* --> a pattern that matches the comma , and matches whitespace 0 or more times, and splits by this pattern by the number specified in the maxsplit parameter
re.split(r',\s*', '1,  2,  3,4,    5,6,7,8', maxsplit=3)
````

### Function findall and finditer—Finding All Matches in a String

````python
contact = 'Wally White, Home: 555-555-1234, Work: 555-555-4321'
````

````python
#\d{3}-\d{3}-\d{4} --> Match exactly 3 digits - Match exactly 3 digits - Match exactly 4 digits
re.findall(r'\d{3}-\d{3}-\d{4}', contact)
````

````python
#\d{3}-\d{3}-\d{4} --> Match exactly 3 digits - Match exactly 3 digits - Match exactly 4 digits
for phone in re.finditer(r'\d{3}-\d{3}-\d{4}', contact):
    print(phone.group())
````

### Capturing Substrings in a Match

````python
text = 'Charlie Cyan, e-mail: demo1@deitel.com'
````

````python
#([A-Z][a-z]+ [A-Z][a-z]+) --> any two string starting with an uppercase Letter, then 1 or more occurrences of lowercase letters
#, --> comma 
# e-mail: (\w+@\w+\.\w{3}) --> Match a single word 1 or more times, then @ then match a single word 1 or more times then . then match a word of three characters 
pattern = r'([A-Z][a-z]+ [A-Z][a-z]+), e-mail: (\w+@\w+\.\w{3})'
````

````python
result = re.search(pattern, text)
````

````python
result.groups()
````

````python
result.group()
````

````python
result.group(1)
````

````python
result.group(2)
````
